```python
import pandas as pd

cols = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy = pd.read_excel('Energy Indicators.xls', 'Energy', 
                       usecols=[2, 3, 4, 5], 
                       names=cols, skiprows=17, 
                       skip_footer=38,
                       na_values=['...'])
```
